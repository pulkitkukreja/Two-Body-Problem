```python
def plot(rs):
    fig = plt.figure()
    ax = fig.add_subplot(111,projection='3d')
    #Unpacking the input matrix
    x1 = rs[:,0]
    y1 = rs[:,1]
    z1 = rs[:,2]
    x2 = rs[:,3]
    y2 = rs[:,4]
    z2 = rs[:,5]
    xG = rs[:,6]
    yG = rs[:,7]
    zG = rs[:,8]
    
    ax.set_xlabel('X(km)')
    ax.set_ylabel('Y(km)')
    ax.set_zlabel('Z(km)')

    #Motion relative to an inertial frame
    
    ax.plot(x1,y1,z1,'w',label="Trajectory of m1")
    ax.plot(xG,yG,zG,'r',label="Trajectory of CM")
    ax.plot(x2,y2,z2,'b',label="Trajectory of m2")
    ax.plot([rs[0,0]],[rs[0,1]],[rs[0,2]],'wo',label='Initial Position of m1')
    ax.plot([rs[0,6]],[rs[0,7]],[rs[0,8]],'ro',label='Initial Position of CM')
    ax.plot([rs[0,3]],[rs[0,4]],[rs[0,5]],'bo',label='Initial Position of m2')
    #Vectors plot
    _x,_y,_z = [[0,0,0],[0,0,0],[0,0,0]]
    _u,_v,_w = [[rs[len(rs[:,0])-1,0],0,0],[0,rs[len(rs[:,1])-1,1],0],[0,0,rs[len(rs[:,2])-1,2]]]
    ax.quiver(_x,_y,_z,_u,_v,_w,color='k')
    ax.set_title('Motion of two bodies as viewed from an inertial frame of reference')
    plt.legend()
    
    #Motion relative to center of mass
    
    fig = plt.figure()
    ax = fig.add_subplot(111,projection='3d')
    ax.plot(x1-xG,y1-yG,z1-zG,'w',label='Trajectory of m1')
    ax.plot(x2-xG,y2-yG,z2-zG,'b',label='Trajectory of m2')
    
    ax.plot([(x1-xG)[0]],[(y1-yG)[0]],[(z1-zG)[0]],'wo',label='Initial Position of m1')
    ax.plot([0],[0],[0],'ro', label='CM')
    ax.plot([(x2-xG)[0]],[(y2-yG)[0]],[(z2-zG)[0]],'bo',label='Initial Position of m2')
    #Vectors plot
    _x,_y,_z = [[0,0,0],[0,0,0],[0,0,0]]
    _u,_v,_w = [[abs((x1-xG)[len(x1-xG)-1]),0,0],[0,abs((y1-yG)[len(y1-yG)-1]),0],[0,0,abs((z1-zG)[len(z1-zG)-1])]]
    ax.quiver(_x,_y,_z,_u,_v,_w,color='k')
    plt.legend()
    ax.set_title("Motion relative to center of mass")
    ax.set_xlabel('X(km)')
    ax.set_ylabel('Y(km)')
    ax.set_zlabel('Z(km)')
    
    #Motion relative mass m1
    
    fig = plt.figure()
    ax = fig.add_subplot(111,projection='3d')
    ax.plot(x2-x1,y2-y1,z2-z1,'b',label='Trajectory of m2')
    ax.plot(xG-x1,yG-y1,zG-z1,'r',label='Trajectory of CM')
    
    ax.plot([(x2-x1)[0]],[(y2-y1)[0]],[(z2-z1)[0]],'bo',label='Initial Position of m2')
    ax.plot([0],[0],[0],'wo', label='m1')
    ax.plot([(xG-x1)[0]],[(yG-y1)[0]],[(zG-z1)[0]],'ro',label='Initial Position of CM')
    #Vectors plot
    _x,_y,_z = [[0,0,0],[0,0,0],[0,0,0]]
    _u,_v,_w = [[abs((x2-x1)[len(x2-x1)-1]),0,0],[0,abs((y2-y1)[len(y2-y1)-1]),0],[0,0,abs((z2-z1)[len(z2-z1)-1])]]
    ax.quiver(_x,_y,_z,_u,_v,_w,color='k')
    
    ax.set_title("Motion relative to mass m1")
    ax.set_xlabel('X(km)')
    ax.set_ylabel('Y(km)')
    ax.set_zlabel('Z(km)')
    
    plt.legend()
    plt.show()
```

```python
#The function that returns the velocity and acceleration 
def diff(t,y):
    #Unpacking the state 
    r1 = np.array([y[0],y[1],y[2]])
    r2 = np.array([y[3],y[4],y[5]])
    
    v1 = [y[6],y[7],y[8]]
    v2 = [y[9],y[10],y[11]]
    
    #norm of r2-r1
    r = np.linalg.norm(r2-r1)
    
    A1 = (G*m2*(r2-r1))/(r**3)
    A2 = (G*m1*(r1-r2))/(r**3)
    
    state_final = v1 + v2 + list(A1) + list(A2)
    return state_final
```

```python
G = 6.67259e-20
#Masses of the bodies
m1 = 1e26
m2 = 1e26

#Initial time
t0 = 0

#Initial position of the masses
r1_0 = [0,0,0]
r2_0 = [3000,0,0]

#Initial velocities of the masses
v1_0 = [10,20,30]
v2_0 = [0,40,0]

#initial state vector
y0 = r1_0 + r2_0 + v1_0 + v2_0
```

```python
#Scipy method of solution of Ordinary Differential Equation by method called 'lsoda'
#The DE solver object
```

```python
solver = ode(diff)
solver.set_integrator('lsoda')
solver.set_initial_value(y0,0)
```

```python
#timespan 
tspan = 1000
#timestep
dt = 1.0
    
n_steps = int(np.ceil((tspan-t0)/dt))
    
#initialze arrays
ys = np.zeros((n_steps,12))
ts = np.zeros((n_steps,1))

ys[0] = np.array(y0)
step = 1
```

```python
#Propagation of the Solution of the DE
while solver.successful() and step<n_steps:
    solver.integrate(solver.t + dt)
    ts[step] = solver.t
    ys[step] = solver.y
    step+=1

x1 = ys[:,0]
y1 = ys[:,1]
z1 = ys[:,2]
x2 = ys[:,3]
y2 = ys[:,4]
z2 = ys[:,5]   

xG = (m1*x1 + m2*x2)/(m1+m2)
yG = (m1*y1 + m2*y2)/(m1+m2)
zG = (m1*z1 + m2*z2)/(m1+m2)


rs = ys[:,:6]


rs = np.column_stack((rs,xG,yG,zG))
plot(rs)
```
